```python
import pandas as pd
```

```python
# Only using capital cities
capitals = {
  'Vienna',
  'Brussels',
  'Sofia',
  'Zagreb',
  'Nicosia',
  'Prague',
  'Copenhagen',
  'Tallinn',
  'Helsinki',
  'Paris',
  'Berlin',
  'Athens',
  'Budapest',
  'Dublin',
  'Rome',
  'Riga',
  'Vilnius',
  'Luxembourg',
  'Valletta',
  'Amsterdam',
  'Warsaw',
  'Lisbon',
  'Bucharest',
  'Bratislava',
  'Ljubljana',
  'Madrid',
  'Stockholm',
}
print(len(capitals))
```

```python
df_all_values = pd.DataFrame()
```

```python
# Air Pollution
df = pd.read_excel(io='../DataSources/RenamedDataSources/ISGlobal-Ranking-table-AP-GS-Noise-23032022.xlsx', 
  sheet_name='ISGlobalCityRanking_AirPollutio')

df_air = df[df['City'].isin(capitals)]
print(df_air['City'].shape)

df_air = df_air.sort_values(by='City', ignore_index=True)
df_all_values['City'] = df_air['City']
df_all_values['PM2.5 [annual mean]'] = df_air['PM2.5 [annual mean]']
df_all_values['NO2 [annual mean]'] = df_air['NO2 [annual mean]']
print(df_all_values)
```

```python
# Noise pollution
df = pd.read_excel(io='../DataSources/RenamedDataSources/ISGlobal-Ranking-table-AP-GS-Noise-23032022.xlsx', 
  sheet_name='ISGlobalCityRanking_Noise')
df['City'].str.strip()

df_noise = df[df['City'].isin(capitals)]

missing_cities = pd.DataFrame([
  {'City':'Nicosia'},
  {'City':'Bucharest'},
  {'City':'Athens'},
  {'City':'Bratislava'},
  {'City':'Zagreb'},
])
df_noise = pd.concat([df_noise, missing_cities])
print(df_noise['City'].shape)

df_noise = df_noise.sort_values(by='City',ignore_index=True)
df_all_values['% population exposed > 55 dB Lden'] = df_noise['% population exposed > 55 dB Lden']
print(df_all_values)
```

```python
# Green Spaces
df = pd.read_excel(io='../DataSources/RenamedDataSources/ISGlobal-Ranking-table-AP-GS-Noise-23032022.xlsx', 
  sheet_name='ISGlobalCityRanking_GreenSpace')

df_green = df[df['City'].isin(capitals)]
print(df_green['City'].shape)

df_green = df_green.sort_values(by='City', ignore_index=True)
df_all_values['%GA [mean level]'] = df_green['%GA [mean level]']
print(df_all_values)
```

```python
# Greenpeace Public Transport Affordability score
df = pd.read_excel(io='../DataSources/RenamedDataSources/GreenPeacePublicTransportAffordability.xlsx', sheet_name="Sheet1")

df_transport_affordability = df[df['City'].isin(capitals)]
print(df_transport_affordability['City'].shape)

df_transport_affordability = df_transport_affordability.sort_values(by='City', ignore_index=True)
df_all_values['Public Transport Affordability'] = df_transport_affordability['Points per city']
print(df_all_values)
```

```python
# EU Commision Quality of Life
df = pd.read_excel(io='../DataSources/RenamedDataSources/quality_of_life_european_cities_2023_aggregated_data.xlsx', 
  sheet_name='QoL in European cities 2023_V2')

new_header = df.iloc[0]
df = df[1:]
df.columns = new_header 

rows_required = {
  # Transport modes
  261,262,263,264,265,266,267,
  # Public transport satisfaction
  33,34,
  # Green spaces satisfaction
  73,74,
  # Air quality satisfaction
  103,104,
  # Noise level satisfaction
  113,114,
  # City satisfaction
  133,134,
}

df_filtered_rows = df.loc[df.index.intersection(rows_required)]
df_questionaire = df_filtered_rows[list(capitals.intersection(set(df_filtered_rows.columns)))]

cities = pd.DataFrame([df_questionaire.columns], columns=df_questionaire.columns)
df_questionaire = pd.concat([cities, df_questionaire], ignore_index=True)

row_labels = [
  'City',
  'Disatisfied Public Transport', 'Satisfied Public Transport',
  'Disatisfied Green Spaces', 'Satisfied Green Spaces',
  'Disatisfied Air Quality', 'Satisfied Air Quality',
  'Disatisfied Noise Level', 'Satisfied Noise Level',
  'Disatisfied City', 'Satisfied City',
  'Car', 'Motorcycle', 'Bicycle', 'Foot', 'Train', 'Bus,Metro,Tram','Other']
df_questionaire.index = row_labels
df_questionaire = df_questionaire.transpose()
df_questionaire = df_questionaire.sort_values(by='City', ignore_index=True)
print(df_questionaire)

df_all_values['Satisfied Public Transport'] = df_questionaire['Disatisfied Public Transport']
df_all_values['Satisfied Green Spaces'] = df_questionaire['Satisfied Green Spaces']
df_all_values['Satisfied Air Quality'] = df_questionaire['Satisfied Air Quality']
df_all_values['Satisfied Noise Level'] = df_questionaire['Satisfied Noise Level']

df_all_values['Satisfied City'] = df_questionaire['Satisfied City']

df_all_values['Private Vehicle'] = df_questionaire['Car'] + df_questionaire['Motorcycle']
df_all_values['Bicycle'] = df_questionaire['Bicycle']
df_all_values['Walk'] = df_questionaire['Foot']
df_all_values['Public Transport'] = df_questionaire['Train'] + df_questionaire['Bus,Metro,Tram']
df_all_values['Other'] = df_questionaire['Other']

print(df_all_values)
```
