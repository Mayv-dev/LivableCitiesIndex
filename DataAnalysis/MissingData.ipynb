```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import KNNImputer
from importnb import Notebook
with Notebook():
  import AllValues

df = AllValues.df_all_values
```

```python
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df.drop(columns=['City']))
print(data_scaled)
```

```python
imputer = KNNImputer(n_neighbors=5)
data_imputed = imputer.fit_transform(data_scaled)

df_complete = pd.DataFrame(data_imputed, columns=df.columns.drop('City'))
df_complete['City'] = df['City']
print(df_complete)
```
