```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from importnb import  Notebook

with Notebook():
  import EnvironmentQuality
  import MobilityAndTransport
  import PublicPerception
  import CitySatisfaction
  import CityLiveability
  
env = EnvironmentQuality.df_environment
trn = MobilityAndTransport.df_transport
per = PublicPerception.df_public
sat = CitySatisfaction.df_city_satisfaction
com = CityLiveability.df_composite
```

```python
city_com = com['City'].to_numpy()
city_env = env['City'].to_numpy()
city_trn = trn['City'].to_numpy()
city_per = per['City'].to_numpy()
city_sat = sat['City'].to_numpy()

```

```python
figure(figsize=(20,20))
x_fix = 30
y = 27
for city in city_com:
  x_vals = []

  x, = np.where(city_com == city)
  x_vals.append(x)

  x, = np.where(city_env == city)
  x_vals.append(x)

  x, = np.where(city_trn == city)
  x_vals.append(x)

  x, = np.where(city_per == city)
  x_vals.append(x)

  x, = np.where(city_sat == city)
  x_vals.append(x)
  
  x_max = x_fix-np.max(x_vals)
  x_min = x_fix-np.min(x_vals)
  
  plt.plot([x_min, x_max],[y, y], 'k')
  
  if y == 27:
    plt.plot(x_fix-x_vals[0],y,color='#777777',markersize=14, marker='o',label='Overall')
    plt.plot(x_fix-x_vals[1],y,color='#2E8B57',markersize=10, marker='o',label='Environment')
    plt.plot(x_fix-x_vals[2],y,color='#4682B4',markersize=10, marker='o',label='Mobility & Transport')
    plt.plot(x_fix-x_vals[3],y,color='#FFA500',markersize=10, marker='o',label='Public Perception')
    plt.plot(x_fix-x_vals[4],y,color='#9370DB',markersize=10, marker='o',label='City Satisfaction')
  else:
    plt.plot(x_fix-x_vals[0],y,color='#777777',markersize=14, marker='o')
    plt.plot(x_fix-x_vals[1],y,color='#2E8B57',markersize=10, marker='o')
    plt.plot(x_fix-x_vals[2],y,color='#4682B4',markersize=10, marker='o')
    plt.plot(x_fix-x_vals[3],y,color='#FFA500',markersize=10, marker='o')
    plt.plot(x_fix-x_vals[4],y,color='#9370DB',markersize=10, marker='o')
    

  plt.text(0,y,city)
  
  y-=1
  
ax = plt.gca()
ax.get_yaxis().set_visible(False)
ax.get_xaxis().set_visible(False)

plt.legend(loc='upper left')

plt.savefig('./plots/overall.png')
plt.show()
```
